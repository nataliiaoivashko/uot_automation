### Question 1 (10 marks):

Create a "Number Guessing" application where the computer picks a number between 1 and 100, and the user has to guess which number the game has given. Each time the guess is incorrect, the application can provide a clue (e.g., divisible by 2, greater than 50, etc.).

```python
import random



def random_num():
    """function that generates random number"""
    random_number = random.randint(1, 100)  # Generating a random number between 1 and 100
    return random_number


def play_game(computer_num=None, user_num=None):
    """Calling the function and storing the computer number in a variable and user number as user input"""
    if computer_num is None:
        computer_num = random_num()

    # print(computer_num)

    if user_num is None:
        user_num = int(input("HELLO! I HAVE PICKED THE NUMBER BETWEEN 1 AND 100. GUESS IT. Enter any number:"))

    hint_counter1 = 0

    while True:
        print(f"Your pick is", {user_num})
        if user_num == computer_num:
            print("You guessed correctly! The number is ", {computer_num})
            print("Guessing Attempts: ", {hint_counter1 + 1})
            break
        if user_num != computer_num:
            hint_counter1 = hint_counter1 + 1
            print("Sorry, try again")
        if user_num <= 0:
            # return str("The number should be greater than zero")
            print("The number should be greater than zero")
        if user_num > 100:
            print("Your number is out of range. Choose between 1 and 100")
        if abs(computer_num - user_num) <= 2:
            print("your guess is extremely close!!!! plus minus 2 (or less) from the guess")
        if abs(computer_num - user_num) <= 10:
            print("You are getting closer: plus minus 10 digits (or less) from the guess")
        if abs(computer_num - user_num) <= 20 and abs(computer_num - user_num) >= 10:
            print("You are getting closer: plus minus 20 digits (or less) from the guess")
        if computer_num % 2 == 0:
            print("Hint: the number is divisible by 2")
        if computer_num % 2 != 0:
            print("Hint: the number is not divisible by 2")
        if computer_num % 3 == 0:
            print("Hint: the number is divisible by 3")
        if computer_num % 3 != 0:
            print("Hint: the number is not divisible by 3")
        if user_num <= 10 and computer_num > 10:
            print("Another Hint: the number is greater than 10")
        elif user_num >= 10 and computer_num < 10:
            print("Another Hint: the number is less than 10")
        elif user_num <= 20 and computer_num > 20:
            print("Another Hint: the number is greater than 20")
        elif user_num >= 20 and computer_num < 20:
            print("Another Hint: the number is less than 20")
        elif user_num <= 30 and computer_num > 30:
            print("Another Hint: the number is greater than 30")
        elif user_num >= 30 and computer_num < 30:
            print("Another Hint: the number is less than 30")
        elif user_num <= 40 and computer_num > 40:
            print("Another Hint: the number is greater than 40")
        elif user_num >= 40 and computer_num < 40:
            print("Another Hint: the number is less than 40")
        elif user_num <= 45 and computer_num > 45:
            print("Another Hint: the number is greater than 45")
        elif user_num >= 45 and computer_num < 45:
            print("Another Hint: the number is less than 45")
        if hint_counter1 > 5:
            if abs(computer_num - user_num) >= 50:
                print("You are too far.")
            if computer_num <= 50 and user_num >= 50:
                print("The guessed number is less than 50")
            elif computer_num >= 50 and user_num <= 50:
                print("The guessed number is greater than 50")
            elif computer_num <= 60 and user_num >= 60:
                print("The guessed number is less than 60")
            elif computer_num >= 60 and user_num <= 60:
                print("The guessed number is greater than 60")
            elif computer_num <= 70 and user_num >= 70:
                print("The guessed number is less than 70")
            elif computer_num >= 70 and user_num <= 70:
                print("The guessed number is greater than 70")
            elif computer_num <= 90 and user_num >= 90:
                print("The guessed number is less than 90")
            elif computer_num >= 90 and user_num <= 90:
                print("The guessed number is greater than 90")
        else:
            print("~~~~~~~~~~~~~~~~~~~~~")
        user_num = int(input("Take another guess:"))
        continue

# to run the game- run the below line
play_game()

# if you need to run the game with pre-set values, run this one line below
#play_game(51, 0)
```

```python
import datetime
from datetime import date
```

```python
def delivery_date(weekday_name, order_date):
    if weekday_name == 'Thursday':
        return order_date + datetime.timedelta(days=4)
    elif weekday_name == 'Friday':
        return order_date + datetime.timedelta(days=4)
    elif weekday_name == 'Saturday':
        return order_date + datetime.timedelta(days=3)
    else:
        return order_date + datetime.timedelta(days=2)
```

```python
updated_orders['delivery_date'] = updated_orders.apply(lambda x: delivery_date(x['weekday_name'], x['order_date']), axis=1)
```

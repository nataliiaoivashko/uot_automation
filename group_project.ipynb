```python
rejections['Rejection_rate'] = rejections['sample_case_rejections1'] + rejections['sample_case_rejections2'] + rejections['sample_case_rejections3'] / rejections['sample_case_count1'] + rejections['sample_case_count2'] + rejections['sample_case_count3']
```

```python
def threshold(x,y,z):
    if x >= 0:
        return f'Hi {z}, you have {x*100}% rejected of {y} products'
```

```python
rejections['Notifications'] = rejections.apply(lambda x: threshold(x['rejection_rate'], x['ingredient_name'], x['supplier']), axis=1)
```
